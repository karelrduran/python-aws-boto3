## Use case 2: Load data into DynamoDB

We are going to load a list of movies into DynamoDB. The first step is to choose the partition and sort keys. The file datosPeliculas.json contains the following content:

The file datosPeliculas.json contains the following content:

```json
[
    {
        "year": 2013,
        "title": "Rush",
        "info": {
            "directors": ["Ron Howard"],
            "release_date": "2013-09-02T00:00:00Z",
            "rating": 8.3,
            "genres": [
                "Action",
                "Biography",
                "Drama",
                "Sport"
            ],
            "image_url": "http://ia.media-imdb.com/images/M/MV5BMTQyM0MTV5TcwMjI2OTI0OQ@@._V1_SX400_.jpg",
            "plot": "A re-creation of the merciless 1970s rivalry between Formula One rivals James Hunt and Niki Lauda.",

            "actors": [
                "Daniel Bruhl",
                "Chris Hemsworth",
                "Olivia Wilde"
            ]
        }
    },
]
```

Since the years of the movies allow us to partition the data in a more or less balanced way, it is the best candidate for the partition key. As there will be several movies in the same year, we choose the title as the sort key, causing the documents to have a composite key.

Therefore, we are going to name our table as FilmsData and set the year attribute of numeric type as the partition key, and the title attribute of string type as the sort key.

<h4 align="center">Creation of the FilmsData table</h4>

Once the table is created, let's see how we can load the data.

````python
import boto3
import os
````

````python
"""
    The credentials are stored into environment variables:
    Las credenciales est√°n almacenadas dentro de variables de entorno
"""

AWS_ACCESS_KEY = os.getenv('AWS_ACCESS_KEY')
AWS_SECRET_KEY = os.getenv('AWS_SECRET_KEY')

AWS_REGION = os.getenv('AWS_REGION')
````

````python
# Load movie data

import json
import decimal

dynamodb = boto3.resource(
    'dynamodb', 
    region_name= AWS_REGION, 
    aws_access_key_id= AWS_ACCESS_KEY, 
    aws_secret_access_key= AWS_SECRET_KEY
)

table = dynamodb.Table('FilmsData')

with open("data/FilmsData.json") as jsonFile:
    movies = json.load(jsonFile, parse_float=decimal.Decimal)
    for movie in movies:
        year = int(movie['year'])
        title = movie['title']
        info = movie['info']


        table.put_item(
            Item={
                'year': year,
                'title': title,
                'info': info,
            }
        )

        print("Movie added:", year, title)
````
